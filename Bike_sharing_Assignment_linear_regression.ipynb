#### Transforming season,weathersit,mnth and weekday columns 

```python
# Transforming weathersit feature
bike_sharing['weathersit'] = bike_sharing['weathersit'].map({1:'Clear to Partly Cloud', 2:'Misty and Cloudy', 3:'Light Rain or Snow', 4: 'Heavy Rain or Snow'})

# Transforming season feature
bike_sharing['season'] = bike_sharing['season'].map({1:'Spring', 2:'Summer', 3:'Fall', 4: 'Winter'})

# Transforming month feature
bike_sharing['mnth'] = bike_sharing['mnth'].map({1:'Jan', 2:'Feb', 3:'Mar', 4: 'Apr', 5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'})

# Transforming weekday feature
bike_sharing['weekday'] = bike_sharing['weekday'].map({0:'Sun', 1:'Mon', 2:'Tue', 3:'Wed', 4: 'Thu',5:'Fri',6:'Sat'})

```

# Model Building

###  Model 1

###### Using Recursive Feature Elimination

```python
# Creating an instance of LinearRegression
lm=LinearRegression()

#fitting model with x_train and y_train

lm.fit(x_train, y_train)

# Creating RFE instance on fitted lm

rfe = RFE(estimator=lm, n_features_to_select=15)

# Converting rfe to fitted rfe

rfe=rfe.fit(x_train, y_train)
```

```python
list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

-This results shows the suggestions given by RFE to retain top 15 features

```python
col=x_train.columns[rfe.support_]
col
```

```python
x_train.columns[~rfe.support_]
```

```python
#Creating x_train dataframe with RFE selected variable
x_train_rfe=x_train[col]
```

## Building Linear Model Using stats model

###  Model 1

VIF Check

Checking the Vif values of the RFE selected columns 

```python


def vif_table(dataframe):
    vif = pd.DataFrame() 
    vif['Feature'] = dataframe.columns
    vif['VIF'] = [variance_inflation_factor(dataframe.values, i) for i in range(dataframe.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by="VIF", ascending=False)
    return vif


result = vif_table(x_train_rfe)
print(result)
```

```python
# add constant

x_train_lm1=sm.add_constant(x_train_rfe)
lr1=sm.OLS(y_train,x_train_lm1).fit()
```

```python
lr1.params
```

```python
lr1.summary()
```

- From the above VIF value it's clear that humidity(hum) and temperature(temp) have high VIF value

- So we can drop the hum column( humidity and temperature are related as our general knowledge, so dropping hum may reduce the VIF value of temp)
